```python
from selenium import webdriver
import csv
import pandas as pd
import re
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
test2 = []
Year = []
publisher = []
y = 1
years = [2017, 2018, 2019, 2020, 2021, 2022]
haikyu = '東宝'

#公開終了の映画を取得
for year in years:
    url = 'https://www.toho.co.jp/movie/archives/{}/index.html'.format(str(year))
    driver.get(url)
    
    while y < 20:
        for x in range(1, 5):              
            aim = 'html/body/div/div[2]/main/div/article/div/div/ul[{0}]/li[{1}]/p'.format(y, x)
                                                                       
            if len(driver.find_elements_by_xpath(aim)) > 0:
                test = driver.find_element_by_xpath(aim)
                test2.append(test.text)
                Year.append(year)
                publisher.append(haikyu)

        y+=1
    y=1                                                                   

#現在公開中の映画を取得
year = 2022
url = 'https://www.toho.co.jp/movie/lineup/index.html'
driver.get(url)


while y < 20:
    for x in range(1, 5):              
        aim = '/html/body/div/div[2]/main/div/article/div/section[1]/div/ul[{0}]/li[{1}]/p'.format(y, x)
                                                                       
        if len(driver.find_elements_by_xpath(aim)) > 0:
            test = driver.find_element_by_xpath(aim)
            test2.append(test.text)
            Year.append(year)
            publisher.append(haikyu)

    y+=1

#dfにする
df = pd.DataFrame(zip(Year, test2, publisher))
df.columns = ['公開年', '映画（作品名）', '配給']
df.to_csv('Kinema_publisher_{}_df_20220528.csv'.format(haikyu))
```
